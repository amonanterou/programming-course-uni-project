```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def run_random_walk(steps):
    """ Σε αυτή τη συνάρτηση εκτελώ έναν τυχαίο περίπατο ανάλογα με τα βήματα που περνάνε
        σαν όρισμα στη συνάρτηση.
    """
    pos = np.array([0, 0]) # αρχική θέση

    positions = [pos] # λίστα με ΄όλες τις θέσεις
    distances = [] # η λίστα με όλες τις αποστάσεις
    
    # για κάθε βήμα ανάλογα με τον αριθμό βημάτων κάνω
    for step in range(steps):
        # επιλέγω τυχαία την κατεύθυνση
        move = rng.choice(rand_moves)
        # κάνω την κίνηση προσθέτοντας το διάνυσμα κατεύθυνσης στην τωρινή μου θέση
        pos = pos + move
        # εισάγω την θέση που πήγα στη λ΄ίστα με όλες τις θέσεις
        positions.append(pos)
        # εισάγω και την απόσταση στην αντίστοιχη λίστα
        distances.append(np.linalg.norm(pos))
    
    # τέλος επιστρέφω όλες τις θέσεις, τη μέγιστη απόσταση, την ελάχιστη απόσταση, καθώς και τη μέση τιμή των αποστάσεων
    return positions, max(distances), min(distances), np.mean(distances)
```

```python
def show_walk(positions):
    # ετοιμάζω τα δεδομένα για να τα εμφανίσω
    x, y = [], []
    for v in positions: # για κάθε διάνυσμα
        x.append(v[0]) # εξάγω την τετμημένη
        y.append(v[1]) # και την τεταγμένη
    
    # plot
    fig, ax = plt.subplots() # παίρνω σε ξεχωριστές μεταβλητές το σχήμα (fig) και το σύστημα αξώνων (ax) που θα εμφανίσω

    ax.plot(x, y, linewidth=2.0) # ζωγραφίζω τα δεδομένα στο επίπεδο
    
    ax.set_title("Τυχαίος Περίπατος")

    plt.show() # εμφανίζω το διάγραμμα

def run_simulation(steps):
    # εκτελώ 100 περιπάτους και εμφανίζω αποτελέσματα
    results = [] # λίστα που αποθηκεύω όλα τα αποτελέσματα
    sim2plot = rng.integers(100) # επιλέγω τυχαία έναν περίπατο από τους 100
    
    # κάνω 100 περιπάτους
    for i in range(100):
        
        # τρέχω τον περίπατω ο οποίος μου επιστρέφει positions, max(distances), min(distances), np.mean(distances)
        r = run_random_walk(steps)
        # εισάγω τα δεδομένα για τις αποστάσεις στη λίστα results
        results.append(r[1:])
        
        # αν πρέπει να εμφανίσω το διάγραμμα
        if i == sim2plot:
            # το εμφανίζω δίνοντας τις θ΄έσεις στην αντίστοιχη συνάρτηση
            show_walk(r[0])
    
    # επιστρέφω τα αποτελέσματα
    return results
```

```python
# επιλέγω αριθμό από βήματα
steps = 10
# φτιάχνω μία λίστα με τα διανύσματα κίνησης (π.χ. με το (0,1) θα κινηθώ μία θέση προς τα θετικά του άξονα y)
rand_moves = [np.array([0, 1]), np.array([0, -1]), np.array([1, 0]), np.array([-1, 0])]
# φτιάχνω ένα αντικείμενο που θα μου δίνει τους τυχαίους αριθμούς
rng = np.random.default_rng()

# τρέχω την προσομοίωση που εκτελεί 100 περιπάτους και μου επιστρέφει τα αποτελ΄έσματα αυτών
r = run_simulation(steps)

# εμφανίζω τα αποτελέσματα
print(np.array(r))
```
